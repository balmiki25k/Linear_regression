#Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#task
1) create a excel file, which has 2 column-month, sales
2) use maths, find the value for oct, nov, dec
3) display the plot with linear line.

#Read the Excel file

```python
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/dataset_files/linear_regression_sales_data.xlsx')
df
```

# Get X and Y

```python
X = df["Month"].values
Y = df["Sales"].values
```

#Use maths to find slope (m) and intercept (b)

```python
x_mean = X.mean()
y_mean = Y.mean()

m = np.sum((X - x_mean) * (Y - y_mean)) / np.sum((X - x_mean) ** 2)
b = y_mean - m * x_mean

# Regression line values for existing X
y_pred_math = m * X + b

# 4) Predict Oct, Nov, Dec (10, 11, 12)
future_months = np.array([10, 11, 12])
future_sales = m * future_months + b

print("Predictions:")
print(f"Oct (10): {future_sales[0]:.2f}")
print(f"Nov (11): {future_sales[1]:.2f}")
print(f"Dec (12): {future_sales[2]:.2f}")

# 5) Plot original data + regression line (+ future predictions)
plt.figure(figsize=(7, 5))
plt.scatter(X, Y, label="Original Data Points")
plt.plot(X, y_pred_math, label="Regression Line (Mathematical)")

# show predicted future points
plt.scatter(future_months, future_sales, label="Predicted (Oct-Dec)")

plt.title("Linear Regression from Scratch")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.legend()
plt.show()
```
